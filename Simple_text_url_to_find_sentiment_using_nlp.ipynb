```python
!pip install newspaper3k
from newspaper import Article
from textblob import TextBlob
import nltk
```

```python
url = 'https://en.wikipedia.org/wiki/Mathematics'              # or replace with text file using
article = Article(url)                                         # with open('mytext.txt', 'r') as f:
article.download()        # to get into script                    text = f.read()
article.parse()           # to make it readable without HTML and all
nltk.download('punkt')
article.nlp()             # preparing it for nlp
```

```python
text = article.text
print(text)
```

```python
text = article.summary          # to take out the unescessary like citaions, reference which every wikepedia has. so Summary
print(text)
```

```python
blob=TextBlob(text)
sentiment = blob.sentiment.polarity # -1 (negative) to 1 (positive)
print(sentiment)      # more towards positive sentiment because maths topic is neutral topic of sucess
```
